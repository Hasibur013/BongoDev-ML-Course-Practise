- One variable => univariate distribution
- Two variable => bivariate distribution

# Bivariate distribution: Joint Probability, P(X, Y)

```python
sns.displot(
    data=penguins, 
    x="bill_length_mm", 
    y="bill_depth_mm",
    kind='hist',
    hue="species",
    col='species',
    # cbar=True
)

""" 
bill_length, bill_depth, species: Adielie, Chinstrap, Gentoo

Data set: bill_length, bill_depth, species

Graph 1: P(bill_length, bill_depth | Adielie)
Graph 2: P(bill_length, bill_depth | Chinstrap)
Graph 3: P(bill_length, bill_depth | Gentoo)

model(bill_length, bill_depth):
    p_adiele = P(bill_length, bill_depth | Adielie)
    p_gentoo = P(bill_length, bill_depth | Gentoo)
    p_chinstrap = P(bill_length, bill_depth | Chinstrap)
    
    max_p_species = max(p_adiele, p_gentoo, p_chinstrap)
    return max_p_species
"""
```

```python
sns.pairplot(penguins)
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```
