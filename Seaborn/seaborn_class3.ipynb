```python
import seaborn as sns

sns.set_style("darkgrid")
sns.__version__
```

```python
tips = sns.load_dataset("tips")

tips.head()
```

```python
titanic = sns.load_dataset("titanic")

titanic.head()
```

```python
diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

```python
tips.value_counts("day")
```

# Strip Plot, kind = 'strip'

```python

sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
)
```

```python
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    jitter=False, # for overlap 
)
```

```python
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    order=["Sun", "Sat", "Thur", "Fri"]
)
```

# Swarm Plot

```python
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    kind="swarm",
)
```

```python
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="smoker",
    col="smoker",
    kind="swarm",
)
```

# Box Plot 
- Statistics: minimum, Q1 (25%), median, Q3 (75%), maximum

```python

sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    kind="box",
)
```

```python
sns.catplot(
    data=tips, 
    x="day", 
    y="total_bill", 
    hue="smoker", 
    kind="box"
)
```

# Boxen plot: Gives more statistics i.e. 12.5%, 25% 50% 75% 87.5% 

```python

sns.catplot(
    data=tips, 
    x="day", 
    y="total_bill", 
    hue="smoker", 
    kind="boxen"
)
```

# Violine plot

```python

sns.catplot(
    data=tips, 
    x="day", 
    y="total_bill", 
    kind="violin"
)
```

# Bar Plot

```python

sns.catplot(
    data=tips, 
    x="day",
    y="total_bill",
    kind="bar"
)
```

# Count plot

```python
sns.catplot(
    data=tips, 
    x="day",
    kind="count"
)
```

# Point Plot

```python
sns.catplot(
    data=titanic,
    x="sex",
    y="survived",
    kind="point"
)
```

```python
sns.catplot(
    data=titanic, 
    x="class", 
    y="survived", 
    hue="sex",
    palette={"male": "g", "female": "m"},
    markers=["^", "o"], 
    linestyles=["dotted", "dashdot"],
    kind="point"
)
```
