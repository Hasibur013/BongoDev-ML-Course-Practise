# Statistical plots in seaborn
  1. Relational plot (relplot)
  2. Distributional plot (displot)
  3. Categorical plot (catplot)

```python
import seaborn as sns

sns.set_style(style="darkgrid")
sns.__version__
```

# Relational plot 
- Represents relation between two variables
- Kind: Scatter (Default), Line

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.info()
```

```python
tips.value_counts(tips["time"])
```

```python
tips.value_counts(tips["day"])
```

# target variable: tip

input variable: total_bill

Hypo 1: tip increase with total_bill

Result: Almost True

```python

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
)
```

# Hypo 2. There may be some relation of tip based on days
Observation: People like to tip more in Saturday/Sunday

```python

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    
)
```

# Hypo 3. People love to spend more on dinner than lunch
- Observation: True

```python

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    col="time",
    style="day",
    row='sex'
)
```

# Line plot

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.info()
```

```python
fmri.value_counts(fmri["region"])
```

```python
sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    hue='region',
    col='region',
    row='event',
    height=3
    # estimator=None,
)
```

```python
fmri.value_counts(fmri["subject"])
```

```python
sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    hue='event',
    col='subject',
    col_wrap=4,
    aspect=0.75,
    linewidth=2.5,
)
```

```python
sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    hue='event',
    row='event',
)
```
